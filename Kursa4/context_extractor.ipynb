Функция создает таблицу, куда записывается глагол(word) и слова\тэги слов\граммемы слов слева и справа от него (L и R соответственно) в пределах одного предложения. Таблица записывается в файл, который мы укажем (filename). 

```python
def extraction(word, table, filename):
    import re, csv
    from time import time
    start = time()
    with open(filename, 'a', encoding='utf-8') as infotable:
        writer = csv.writer(infotable, quoting=csv.QUOTE_NONE, escapechar='', quotechar='', delimiter='\t', lineterminator='\n')
        col_names = ['L10','L9','L8','L7','L6','L5','L4','L3','L2','L1','Verb','R1','R2','R3','R4','R5','R6','R7','R8','R9','R10']
        writer.writerow(col_names)
        infotable.flush()
        for index, row in table.iterrows():
            if row['lemma'] == word:
                line  = [''] * 21
                line[10] = word
                words = 0
                left = 1
                while table['gramm'].loc[index - left] != 'SENT' and words < 10 and index - left >= 0:
                    if re.search('\w', table['token'].loc[index - left]):
                        words += 1
                        line[10-words] = table['lemma'].loc[index-left]
                    left += 1
                right = 1
                words = 0
                while table['gramm'].loc[index + right] != 'SENT' and words < 10 and index + right < len(table):
                    if re.search('\w', table['token'].loc[index + right]):
                        words += 1
                        line[10 + words] = table['lemma'].loc[index + right]
                    right +=1
                writer.writerow(line)
                infotable.flush()
    print('Extraction time: {}'.format((time() - start) / 60))
```
