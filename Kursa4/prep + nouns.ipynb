```python
import pandas as pd
```

```python
def prepfinder(word, table, filename):
    
    import re, csv
    from time import time
    start = time()
    
    cases = {'n': 'nominative', 'g': 'genitive', 'd': 'dative', 'a': 'accusative', 'v': 'vocative', 'l': 'locative', 
             'i': 'instrumental'}
    
    prep_counter = dict()
    noun_counter = dict()
    cases_counter = dict()
    
    with open(filename, 'a', encoding='utf-8') as infotable:
        writer = csv.writer(infotable, quoting=csv.QUOTE_NONE, escapechar='', quotechar='', delimiter='\t', lineterminator='\n')
        col_names = ['L10','L9','L8','L7','L6','L5','L4','L3','L2','L1','Verb','R1','R2','R3','R4','R5','R6','R7','R8','R9','R10']
        writer.writerow(col_names)
        infotable.flush()
        for index, row in table.iterrows():
            if row['lemma'] == word:
                tags  = [''] * 21
                lemmas = [''] * 21
                tags[10] = 'V'
                lemmas[10] = word
                words = 0
                left = 1
                while table['gramm'].loc[index - left] != 'SENT' and words < 10 and index - left >= 0:
                    if re.search('\w', table['token'].loc[index - left]):
                        words += 1
                        if table['POS'].loc[index - left] in 'NS':
                            tags[10 - words] = table['gramm'].loc[index-left]
                            lemmas[10 - words] = table['lemma'].loc[index-left]
                            current_word = table['lemma'].loc[index-left]
                            current_gramm = table['gramm'].loc[index-left]
                            if table['POS'].loc[index - left] == 'N' and len(current_gramm) >= 5:
                                noun_counter[current_word] = noun_counter.get(current_word, 0) + 1
                                cases_counter[current_gramm[4]] = cases_counter.get(current_gramm[4], 0) + 1
                            else:
                                prep_counter[current_word] = prep_counter.get(current_word, 0) + 1
                            
                        else:
                            tags[10 - words] = '-'
                            lemmas[10 - words] = '-'
                    left += 1
                
                right = 1
                words = 0
                while table['gramm'].loc[index + right] != 'SENT' and words < 10 and index + right < len(table):
                    if re.search('\w', table['token'].loc[index + right]):
                        words += 1
                        if table['POS'].loc[index + right] in 'NS':
                            tags[10 + words] = table['gramm'].loc[index + right]
                            lemmas[10 + words] = table['lemma'].loc[index + right]
                            current_word = table['lemma'].loc[index + right]
                            current_gramm = table['gramm'].loc[index + right]
                            if table['POS'].loc[index + right] == 'N' and len(current_gramm) >= 5:
                                noun_counter[current_word] = noun_counter.get(current_word, 0) + 1
                                cases_counter[current_gramm[4]] = cases_counter.get(current_gramm[4], 0) + 1
                            else:
                                prep_counter[current_word] = prep_counter.get(current_word, 0) + 1
                        else:
                            tags[10 + words] = '-'
                            lemmas[10 + words] = '-'
                    right +=1
                
                writer.writerow(tags)
                writer.writerow(lemmas)
                writer.writerow(' ')
                infotable.flush()
                
    print('Extraction time: {}'.format((time() - start) / 60))
    return (prep_counter, noun_counter, cases_counter)
```

```python
document = pd.read_csv('.\corpora\corpus-i-part1.txt', delimiter='\t', header = 0)
```

```python
(preps, nouns, cases) = prepfinder('заменить', document, 'prep+nouns.csv')
```
